```python
class minHeap:
    def __init__(self):
        self._arr = []
        self._maxSize = 0
        self._size = 0
        
    def insertArr(self, arr):
        self._maxSize = len(arr)
        for i in arr:
            self.insert(i)
            
        return
    
    # pubble-up
    def insert(self, v):
        if (self.isFull()):
            return False
        
        self._arr.append(v)
        hole = self._size   # size = index + 1
        self._size += 1
        while (hole > 0) & (v < self._arr[(hole - 1) // 2]):
            self.swap(hole, (hole - 1) // 2)
            hole = (hole - 1) // 2
            
        return True
    
    def extractMin(self):
        if (self.isEmpty()):
            return False
        
        # replace root with last node
        # and remove the origin root
        self._size -= 1
        self.swap(0, self._size)
        root = self._arr.pop(self._size)
        
        #Heapify 
        self.minHeapify(0)
        
        return root 
    
    # pubble-down
    def minHeapify(self, i):
        to_swap = i
        left_child = 2 * i + 1
        right_child = 2 * (i + 1)
        
        # make sure child exists
        # and find the smallest of the childs
        if (left_child < self._size):
            if (right_child < self._size):  # 2 childs
                # left prior to right if left = right
                # if parent larger than the smaller child
                # then swap them
                
                # left > right and right > parent
                if (self._arr[left_child] > self._arr[right_child]):
                    if (self._arr[i] > self._arr[right_child]):
                        to_swap = right_child
                
                # left <= right and left > parent
                elif (self._arr[i] > self._arr[left_child]):
                    to_swap = left_child
                 
            # 1 child (left guaranteed since 
            # there shouldn't be any holes in a balanced tree)        
            elif (self._arr[left_child] < self._arr[i]):
                to_swap = left_child
                
        
            
        if (to_swap != i):
            self.swap(i, to_swap)
            self.minHeapify(to_swap)
        
    
    def isFull(self):
        return (self._size == self._maxSize)
    
    def isEmpty(self):
        return self._size == 0
    
    def dispArr(self):
        print(self._arr)
        return
    
    def swap(self, i1, i2):
        try:
            self._arr[i1], self._arr[i2] = self._arr[i2], self._arr[i1]
        except:
            return False
        return True
    
    # sorts form large to small
    def heapSort(self):
        while not (self.isEmpty()):
            min = self.extractMin()
            self._arr.append(min)
        return;
    
    
```
